```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#left_lines = open('NaozFabrycky2014_left.txt').readlines()
left_lines = open('NaozFabrycky2014_merger.txt').readlines()
i1_left = []
e1_left = []
e2_left = []
a1_left = []
beta_left = []
#print(left_lines[0].split('\t'))
for l in left_lines:
    t = l.split('\t')
    i1_left.append(float(t[5]))
    e1_left.append(1-float(t[1]))
    e2_left.append(1-float(t[2]))
    a1_left.append(float(t[4].split('  ')[1]))
    beta_left.append(float(t[10]))
    
peri_left = np.array(a1_left)*np.array(e1_left)
```

```python
#right_lines = open('NaozFabrycky2014_right.txt').readlines()
right_lines = open('NaozFabrycky2014_dissipation.txt').readlines()
i1_right = []
e1_right = []
e2_right = []
a1_right = []
beta_right = []
for l in right_lines:
    t = l.split('\t')
    i1_right.append(float(t[5]))
    e1_right.append(1-float(t[1]))
    e2_right.append(1-float(t[2]))
    a1_right.append(float(t[4].split('  ')[1]))
    beta_right.append(float(t[10]))
    
peri_right = np.array(a1_right)*np.array(e1_right)
```

```python
fig = plt.figure(figsize=(4,5))

ax1 = fig.add_axes([0.1, 0.7, 0.8, 0.4],
                   xticklabels=[], ylim=(0, 180))
ax1.plot(i1_left, linewidth=0.5,color='red')
ax1.plot(beta_left, linewidth=0.95, color='limegreen')
ax1.set_ylabel('$i_1$,$\Psi_1$ [deg]')
ax1.set_title('Merger')
ax1.set_ylim(0,175)

ax2 = fig.add_axes([0.1, 0.3, 0.8, 0.4])
ax2.plot(e1_left, linewidth=0.5,color='red')
ax2.plot(e2_left, linewidth=0.5, color='cyan')
ax2.set_ylim(1e-4,1)
ax2.set_yscale('log')
ax2.set_ylabel('1-e1')

ax3 = fig.add_axes([0.1, -0.07, 0.8, 0.4])
ax3.plot(a1_left, linewidth=0.5,color='red')
ax3.plot(peri_left, linewidth=0.5, color='blue')
ax3.set_yscale('log');ax3.set_yticks([0.01, 0.1,1,10,10**2,10**3,10**4])
ax3.set_ylabel('[AU]');ax3.set_ylim(0.001,10**4.)
#ax3.set_xticks('[]')


ax4 = fig.add_axes([0.9, 0.7, 0.8, 0.4],
                  yticklabels=[], ylim=(0, 180))
ax4.plot(i1_right, linewidth=0.5,color='red')
ax4.plot(beta_right, linewidth=0.5, color='limegreen')
ax4.set_title('Tidal Dissipation')
#ax4.set_xlim(0,80000)
ax4.set_xticks([])
ax4.set_ylim(0,175)

ax5 = fig.add_axes([0.9, 0.3, 0.8, 0.4])
ax5.plot(e1_right, linewidth=0.5,color='red')
ax5.plot(e2_right, linewidth=0.95, color='cyan')
ax5.set_ylim(1e-4,1)
ax5.set_yscale('log')
ax5.set_yticklabels([]);ax4.set_xticks([])
#ax5.set_xlim(0,80000)

ax6 = fig.add_axes([0.9, -0.07, 0.8, 0.4],
                  yticklabels=[])
ax6.plot(a1_right, linewidth=0.5,color='red')
ax6.plot(peri_right, linewidth=0.5, color='blue')
ax6.set_yscale('log')
ax6.set_ylim(0.001,10**4.)
ax6.set_yscale('log');ax3.set_yticks([0.01, 0.1,1,10,10**2,10**3,10**4])
ax6.set_yticklabels([])
#ax6.set_xlim(0,80000)
```
