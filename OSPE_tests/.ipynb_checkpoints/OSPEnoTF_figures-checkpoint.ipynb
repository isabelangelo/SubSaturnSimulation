```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines_noGR = open('fig3_noGR.txt').readlines()
i1_noGR = []
e1_noGR = []
for l in lines_noGR:
    t = l.split('\t')
    i1_noGR.append(float(t[5]))
    e1_noGR.append(1-float(t[1]))
```

```python
fig = plt.figure(figsize=(4,5))
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(0, 180))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4])
ax1.plot(i1_noGR, linewidth=0.5,color='b')
ax1.set_ylabel('i1 [deg]')
ax2.plot(e1_noGR, linewidth=0.5,color='b')
ax2.set_yscale('log')
ax2.set_ylabel('1-e1')
```
