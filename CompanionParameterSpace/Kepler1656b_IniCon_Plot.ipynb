```python
# generate large number of initial conditions
from generate_inicon import *
a2vals=[];m3vals=[]
a1vals=[]
for i in range(1000):
    inicons=initial_conditions()
    a2vals.append(inicons[-7])
    m3vals.append(inicons[2])
    a1vals.append(inicons[-8])
```

```python
# overplot initial conditions onto parameter space plot
from Kepler1656_companion_parameter_space import *

# plot mass, separation (AU)
fig,ax=plt.subplots(figsize=(10,6))

# AO data
ax.fill_between(AO_sma,AO_mass,10,alpha=0.3,color='lightpink', hatch='////',
                label='Keck AO')

# Speckle data
ax.fill_between(Speckle_sma,Speckle_mass,10,alpha=0.3,color='coral', hatch='////',
               label='Speckle')

# GAIA data
ax.fill_between([GAIA_sep, 1e5], [GAIA_mass,GAIA_mass], 10, alpha=0.3, color='khaki', hatch='///',
               label='Astrometry')

# RV data
ax.fill_between(RV_sma,RV_mass,10,alpha=0.3,color='lightblue', hatch='///',
               label='Radial Velocities')

# plot Kepler-1656b
ax.plot(0.197, 48.6*Mearth/Msun, 'y.', markersize=10, markeredgecolor='k')
ax.annotate('Kepler-1656b',(0.2,3e-4))
# set labels
ax.set_xlabel('separation (au)',fontsize=12,family='serif')
ax.set_ylabel('companion mass ($M_\odot$)',fontsize=12,family='serif')
ax.legend(loc='upper left')

# plot simulation initial conditions
ax.plot(a2vals,m3vals,'.',color='rosybrown')
# limits and scales
ax.set_xscale('log')
ax.set_xlim(0.1,1e5)
ax.set_yscale('log')
ax.set_ylim(1e-7,10)

# annotate regions from theory
ax.text(0.2,3e-5,r'$a_{outer} > 10a_{inner}$')
#ax.text(3,3e-6,r'$m_{outer} > 0.1m_{inner}$')
ax.text(10,5e-6,r'$m_{outer} < 0.1m_{inner}, a_{outer} >30$ : no EKL')
ax.text(1.6e4,1e-2,r'$a_{outer} < 10^4$au')

# second x axis for separation
ax2 = ax.twiny()
ax2.plot(au2arcsec(AO_sma, system_distance), 3*np.ones(len(AO_sma)), alpha=0)
ax2.tick_params(axis='x', direction='in', which='both')
ax2.set_xscale('log')
ax2.set_xlabel('separation (arcsec)', fontsize=12, family='serif')
```
