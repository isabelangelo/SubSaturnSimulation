```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lines = open('test6.txt').readlines()
```

```python
time = []
e1 = []
i1 = []
e2 = []
for line in lines:
    l = line.split('\t')
    time.append(float(l[2])) 
    e1.append(float(l[3]))
    e2.append(float(l[4]))
    i1.append(float(l[8]))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121);plt.plot(time,e1);plt.xlabel('time');plt.title('e1')
plt.subplot(122);plt.plot(time,e2);plt.xlabel('time');plt.title('e2')
```
