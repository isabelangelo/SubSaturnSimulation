```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# generate a2 for planet case
def compute_planet_a2(e2, a1):
    epsilon = 0.1
    
    a2_found=0;countm=0;Mmax=500
    while a2_found==0:
        a2_eps = (a1/epsilon)*e2/(1-e2**2.)
        a2_max = 150 # where does this 150 come from?
        if a2_max>a2_eps:
            a2_found=1
            a2_min = max(4,a2_eps)
            a2 = np.random.uniform(a2_min,a2_max)
            return a2
        else:
            countm+=1
            if countm==Mmax:
                print('couldn\'t find stable system')
                return None # reject for unstable systems
            
```

```python
# generate a2 for stellar case (see Alexander's code)
def compute_stellar_a2(m1, m2, m3, e1, e2, a1):
    
    a2_found=0;countm=0;Mmax=500
    while a2_found ==0:
    
        # logP from gaussian (mu=0,std=1)
        mu=0;sigma=1;logPval=np.random.normal(mu,sigma,1)[0]
        
        # scale to new mean+variance
        siglogP=2.3;logPBAR=4.8;logP=logPval*siglogP+logPBAR
        
        # convert to P value in years
        daytoyr = 0.00273973;P2=(10**logP)*daytoyr;
        
        # convert to a2
        a2 = (P2**2.*(m1+m2+m3))**(1/3.)
        epsilon = (a1/a2)*e2/(1.-e2**2.)
        
        # check if the system is stable
        if (epsilon<0.1) and (a2<10000.) and ((a1/a2)<0.1):
            a2_found=1
            return a2 # return for stable systems
        elif ((epsilon >= 0.1) or (a2 > 10000) or ((a1/a2)>=0.1)):
            countm+=1
            if countm==Mmax:
                print('couldn\'t find stable system')
                return None # reject for unstable systems
```

# Generate Initial Conditions

```python
def initial_conditions(m3 = 9.548e-2):
    
    # compute masses [Msun]
    m1 = 1.03
    m2 = 1.4585e-4 #48.6Mearth

    # compute radii [Rsun]
    R1 = 1.1
    R2 = 4.594e-2 #5.02Rearth

    # compute spins [fraction of day]
    spin1P = 10 
    spin2P = 10 

    # compute betas [deg]
    beta = 0
    beta2 = 0

    # compute gamma [deg]
    gamma = 45
    gamma2 = 45 
    
    # generate g values [deg]
    g1 = np.random.uniform(0,360)
    g2 = np.random.uniform(0,360)

    # compute mutual inclination [deg]
    cosi = np.random.uniform(np.cos(0),np.cos(np.pi))
    i = np.rad2deg(np.arccos(cosi))

    # compute age [Myr]
    age =  6.31*1.2*1000 #t_age=6.31Gyr, print at 0.8t_age, t_age, 1.2t_age
    
    # compute eccentricities and sma
    stable_system_found = False
    while stable_system_found == False:
        # compute eccentricities
        e1 = 0.01
        e2 = np.random.uniform(0,1)
        # compute semi-major axes [AU]
        a1 = np.random.uniform(1,10) # current value=0.197 AU
        if m3>=0.1: # star case
            a2 = compute_stellar_a2(m1, m2, m3, e1, e2, a1)
        else: # planet case
            a2 = compute_planet_a2(e2, a1)
            
        if a2 is not None:
            stable_system_found=True
  
    return(np.array([m1,m2,m3,R1,R2,spin1P,spin2P,beta,beta2,gamma,gamma2,a1,a2,e1,e2,g1,g2,i,age]))
```

## Initial Condition Distributions

fixed: m1, m2, R1, R2, spin2, spin2, beta, beta2, gamma, gamma2, e1, age

sampled: m3(?), a1, a2(?), e2, g1, g2, i

```python
# generate sample parameters
num_samples = 10000
#samples_stellar = np.empty([num_samples, len(initial_conditions())])
samples_planet = np.empty([num_samples, len(initial_conditions())])
for i in range(num_samples):
#    samples_stellar[i,:] = initial_conditions(m3=0.1)
    samples_planet[i,:] = initial_conditions(m3=9.548e-4)
```
