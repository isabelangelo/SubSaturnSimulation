```python
# to do : fix print statment that leads to %m chance :-4 to :-1 in object

import numpy as np
import matplotlib.pyplot as plt
```

```python
# define run directory

run_dir = '../SimulationResults/Run4/'

# define file directory
file_dir = 'TimeSeries/output_timeseries_'
```

```python
class TimeSeries(object):
    
    def __init__(self, n):
        self.model_number = n
        self.filepath = run_dir + file_dir + str(n) + '.txt'
        # need to store time series
        lines = open(self.filepath).readlines()
        time=[]
        a1=[];e1=[];i1=[]
        a2=[];e2=[];i2=[]
        for line in lines:
            l = [float(x) for x in line.split('\t')[:-4]]
            time.append(l[2])

            a1.append(l[7])
            e1.append(l[3])
            i1.append(l[8])

            a2.append(l[25])
            e2.append(l[4])
            i2.append(l[9])
            
            self.time = np.array(time)
            self.a1 = np.array(a1)
            self.e1 = np.array(e1)
            self.i1 = np.array(i1)
            self.a2 = np.array(a2)
            self.e2 = np.array(e2)
            self.i2 = np.array(i2)
            
    def plot(self):
        fig = plt.figure(figsize=(9,10))

        ax1 = fig.add_subplot(321)
        ax1.plot(self.time,self.i1,color='limegreen')
        ax1.set_ylabel('i (deg)')
        ax1.set_title('Planet b')


        ax2 = fig.add_subplot(323)
        ax2.plot(self.time,self.e1,color='red')
        ax2.set_ylabel('e')
        ax2.set_ylim(-0.1,1)

        ax3 = fig.add_subplot(325)
        ax3.plot(self.time,self.a1,color='blue')
        ax3.set_ylabel('a (au)')
        ax3.set_xlabel('t (yr)')
        ax3.set_ylim(np.min(self.a1)-0.1,np.max(self.a1)+1)

        ax4 = fig.add_subplot(322)
        ax4.plot(self.time,self.i2,color='limegreen')
        ax4.yaxis.tick_right()
        ax4.set_title('Planet c')

        ax5 = fig.add_subplot(324)
        ax5.plot(self.time,self.e2,color='red')
        ax5.yaxis.tick_right()
        ax5.set_ylim(-0.1,1)

        ax6 = fig.add_subplot(326)
        ax6.plot(self.time,self.a2,color='blue')
        ax6.set_xlabel('t (yr)')
        ax6.yaxis.tick_right()
        ax6.set_ylim(np.min(self.a2)-0.1,np.max(self.a2)+1)
        
        plt.savefig('./timeseries_'+str(self.model_number)+'.png')
        
```

```python
for i in (6,4,16,17,25,44,70,73,82,89,94,96):
    TimeSeries(i).plot()
```
