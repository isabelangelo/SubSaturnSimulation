```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = open('Kepler1656_test8.txt').readlines()
t = []
i1 = []
e1 = []
e2 = []
a1 = []
a2 = []
beta = []

for line in lines:
    l = [float(x) for x in line.split('\t')[:-1]]
    t.append(l[2])
    a1.append(l[7])
    a2.append(l[-10])
    e1.append(l[3])
    e2.append(l[4])
    i1.append(l[10])
```

```python
plt.figure(figsize=(19,10))
plt.subplot(232);plt.plot(t,e1);plt.ylabel('e1');plt.xlabel('t[yr]')#[1-i for i in e1])
plt.subplot(235);plt.plot(t,e2);plt.ylabel('e2');plt.xlabel('t[yr]')
plt.subplot(231);plt.plot(t,a1);plt.ylabel('a1');plt.xlabel('t[yr]')
plt.subplot(234);plt.plot(t,a2);plt.ylabel('a2');plt.xlabel('t[yr]')
plt.subplot(233);plt.plot(t,i1);plt.ylabel('i1');plt.xlabel('t[yr]')
```
