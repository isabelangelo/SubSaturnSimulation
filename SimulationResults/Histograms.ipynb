```python
from models import *
from sort_OSPE_output import *
pd.set_option('display.max_rows', None)

starroche = np.transpose([int(i[19:-4]) for i in zeros]) # inner planet tidally disrupted by star
survived = np.transpose([int(i[19:-4]) for i in ones_and_twos])
planetroche = np.transpose([int(i[19:-4]) for i in threes]) # inner planet tidally disrupted by outer planet
failed = np.transpose([int(i[19:-4]) for i in negatives])

crashed = [63, 4, 53, 60, 156, 198, 135, 132, 103, 10, 11, 108, 174, 200, 35, 6, 56]
negative = [2, 86, 92, 107, 128,  149, 154, 170, 171, 178, 182, 199]
```

```python
planet_a = [];companion_a = []
planet_e = [];companion_e = []

planet_aroche = [];companion_aroche = []
planet_eroche = [];companion_eroche = []

planet_asurvived = [];companion_asurvived = []
planet_esurvived = [];companion_esurvived = []

for n in range(1,401):
    if n in survived or n in starroche:# or n in planetroche:
        if n not in negative:
            m=Model(n)
            planet_a.append(m.a1[-1])
            companion_a.append(m.a2[-1])
            planet_e.append(m.e1[-1])
            companion_e.append(m.e2[-1])
        if n in starroche:
            planet_aroche.append(m.a1[-1])
            companion_aroche.append(m.a2[-1])
            planet_eroche.append(m.e1[-1])
            companion_eroche.append(m.e2[-1])
            
        if n in survived:
            planet_asurvived.append(m.a1[-1])
            companion_asurvived.append(m.a2[-1])
            planet_esurvived.append(m.e1[-1])
            companion_esurvived.append(m.e2[-1])
            
            
```

```python
fig,ax=plt.subplots(2,2,figsize=(12, 10))

ax[0,0].hist(planet_a,bins=10,histtype='step',color='dodgerblue',linewidth=1.5,density=True,label='total')
ax[0,0].set_xlabel('a1 (au)',fontsize=15,family='serif')

ax[0,1].hist(companion_a,bins=10,histtype='step',color='dodgerblue',linewidth=1.5,density=True)
ax[0,1].set_xlabel('a2 (au)',fontsize=15,family='serif')

ax[1,0].hist(planet_e,bins=10,histtype='step',color='dodgerblue',linewidth=1.5,density=True)
ax[1,0].set_xlabel('e1',fontsize=15,family='serif')

ax[1,1].hist(companion_e,bins=10,histtype='step',color='dodgerblue',linewidth=1.5,density=True)
ax[1,1].set_xlabel('e2',fontsize=15,family='serif')
plt.legend()
```
